# Binary Classification -- "Least Concern" vs. Others

The simplest way to look at this data is through a binary classification between the "Least Concern" class and all others that reflect some degree of vulnerability to extinction. This approach is common in published studies that have examined this problem, in part due to greater class balance than would be found in multi-class approaches.

As an opporunity to explore more of the XGBoost architecture, the following code makes use of gradient-boosted methods. 

This approach also presents an opportunity to evaluate imputation methods. XGBoost has default approaches to missing observations which can be compared as a "naive" counterpart to the previously generated imputed dataset.

```python
import pandas as pd
pd.options.mode.chained_assignment = None
import numpy as np
from matplotlib import pyplot as plt
import copy
from plotnine import *
from xgboost import XGBClassifier

from sklearn.pipeline import make_pipeline
from sklearn.pipeline import make_pipeline
from sklearn.metrics import roc_auc_score, roc_curve, auc, plot_roc_curve
from sklearn.metrics import precision_recall_curve
from sklearn.metrics import plot_precision_recall_curve
from sklearn.model_selection import StratifiedKFold
from sklearn.model_selection import RandomizedSearchCV
```

```python
# Reading in feature datasets
features_std = pd.read_csv("../data/features_std.csv", index_col=0)
features_imp = pd.read_csv("../data/features_imputed.csv", index_col=0)

# Defining response, cleaning non-imputed data
y_full = features_std["Red List status"]
features_std.drop(["SpecCode", "Genus","Species", "Family", 
                           "Order", "Class", "Red List status", "Speccode"],
                          axis = 1,
                 inplace = True)
```

## Classification with missing data

To begin, an XGBoost classifier will be tuned and evaluated using stratified K fold cross validation. While a full grid search of different hyperparameters would be useful, a randomized search of a subset was run given the large number of potential hyperparameter combinations 

```python
x_full = features_std # Standardized predictors
y_full = y_full == "LC" * 1 # Classes - whether "Least Concern" or not 
y_full = 1 - y_full
```

```python
params = {
        'min_child_weight': [1, 5, 10],
        'gamma': [0.5, 1, 1.5, 2, 5],
        'subsample': [0.6, 0.8, 1.0],
        'colsample_bytree': [0.6, 0.8, 1.0],
        'max_depth': [3, 4, 5, 8, 10, 20],
        'n_estimators': [100,200,300,500,600,800,1000]
        }

xgb_classifier = XGBClassifier(learning_rate=1,
                    objective='binary:logistic',
                    silent=True, 
                    nthread=1)

folds = 5
param_comb = 25

skf = StratifiedKFold(n_splits=folds, shuffle = True, random_state = 0)
```

```python
# Initializing cross-validation search
random_search = RandomizedSearchCV(xgb_classifier, param_distributions=params, 
                                   n_iter=param_comb, scoring='accuracy', n_jobs=4, 
                                   cv=skf.split(x_full,y_full), verbose=3, random_state=0)

# Running randomized CV search
random_search.fit(x_full, y_full)

random_search.best_score_
```

```python
# Adapted in part from https://scikit-learn.org/stable/auto_examples/model_selection/plot_roc_crossval.html#sphx-glr-auto-examples-model-selection-plot-roc-crossval-py
tprs = []
aucs = []
mean_fpr = np.linspace(0, 1, 100)
x_full = x_full.to_numpy()
bestmod_1 = copy.copy(random_search.best_estimator_)

fig, ax = plt.subplots()
for i, (train, test) in enumerate(skf.split(x_full,y_full)):
    bestmod_1.fit(x_full[train], y_full[train])
    viz = plot_roc_curve(bestmod_1, x_full[test], y_full[test],
                         name='ROC fold {}'.format(i),
                         alpha=0.3, lw=1, ax=ax)
    interp_tpr = np.interp(mean_fpr, viz.fpr, viz.tpr)
    interp_tpr[0] = 0.0
    tprs.append(interp_tpr)
    aucs.append(viz.roc_auc)
    
mean_tpr = np.mean(tprs, axis=0)
mean_auc = auc(mean_fpr, mean_tpr)
mean_tpr[-1] = 1.0
std_auc = np.std(aucs)

ax.plot(mean_fpr, mean_tpr, color='b',
        label=r'Mean ROC (AUC = %0.2f $\pm$ %0.2f)' % (mean_auc, std_auc),
        lw=2, alpha=.8)
mean_auc = auc(mean_fpr, mean_tpr)
```

Accuracy on test data is quite high, ~0.77. However, the unbalanced nature of the response makes accuracy / AUC difficult to interpret -- labelling every class as "LC" would attain roughly 70% accuracy. 

Instead, use of a precision-recall curve to highlight TPR / FPR rates may prove more meaningful.

```python
fig, ax = plt.subplots()

for i, (train, test) in enumerate(skf.split(x_full,y_full)):
    bestmod_1.fit(x_full[train], y_full[train])
    viz = plot_precision_recall_curve(bestmod_1, x_full[test], y_full[test],
                         name='P-R fold {}'.format(i),
                         alpha=0.3, lw=1, ax=ax)

ax.axhline(y = sum(y_full == 1) / len(y_full))
```

Performance here seems good, though not fantastic. Precision remains high at low recall, though decays quickly towards a no-skill classifier (highlighted as a blue line) at intermediate recall.

Overall model accuracy does not seem to change substantially with imputation.

```python
fig, ax = plt.subplots()

for i, (train, test) in enumerate(skf.split(x_full,y_full)):
    bestmod_1.fit(x_full[train], y_full[train])
    viz = plot_precision_recall_curve(bestmod_1, x_full[test], y_full[test],
                         name='P-R fold {}'.format(i),
                         alpha=0.3, lw=1, ax=ax)

ax.axhline(y = sum(y_full == 1) / len(y_full))
```

P-R curves show some improvement, however. Average precision has increased, and the general shape of the P-R curves suggests better precision at intermediate recall.

## Variable Importance Plots

Variable importance plots, which quantify the relative contribution of different variables to the splits within each classifier, suggest which features contribute most to classification.

Here, I plot the 15 most important variables, including missingness indicators. Two missing variables, freshwaterMissing and LengthMissing, appear to be important features. There is no clear indication why this might be the case, save that species with fewer known characteristics may be more likely to be labelled "LC".

Other features, on the other hand, make intuitive sense. Whether a species is characterized as cave-dwelling ("cave"), for example, is likely to be an important indicator, as many cave-dwelling fish are threatened with extinction due to habitat loss and specialized life history strategies. The same may be said for freshwater species ("Lakes", "Stream"), and likely the inverse of ocean-dwelling ("Oceanic") species. Larger fish (related to "length") are also likely to be more threatened by extinction. 

```python
imputed_cols = features_std.columns.append(features_std.columns[features_std.isnull().any()] + "Missing")
var_imp = pd.concat([pd.Series(bestmod_2.feature_importances_),
             pd.Series(imputed_cols)],axis = 1)
var_imp.columns=["importance", "variable"]
var_imp.sort_values('importance')

(ggplot(aes(y = 'importance',
            x = 'variable'),
       (var_imp.sort_values('importance', ascending=False))[:15]) +
 geom_bar(stat = "identity") +
 coord_flip())
```

Removing these missingness indicators from this same plot shows few new patterns.
